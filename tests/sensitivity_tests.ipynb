Тесты работоспособности функции нахождения чувствительности для наборов содержащих рост и вес людей из выборки.

```python
import sensitivity as s
import numpy as np
```

```python
# Средние параметры k и b зависимости веса от роста
k = 0.5
b = 10
```

```python
n_sample = 150

X = np.random.normal(170, 10, n_sample).reshape(-1, 1)
y = k*X + b + np.random.normal(0, 3, n_sample).reshape(-1, 1)

k_sens, b_sens = s.getsensitivity(X, y, all_graphs = False, last = True)
```
